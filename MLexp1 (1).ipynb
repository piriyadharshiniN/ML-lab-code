```python
import pandas as pd
import matplotlib.pyplot as plt
print("NAME: Piriyadharshini N")
print("ROLL NO: 24BAD086")

# Load Dataset
df = pd.read_csv("diabetes.csv")
# Dataset Information
print("\nDataset Info:")
print(df.info())

print("\nMissing Values Before Handling:")
print(df.isnull().sum())
#replacing 0 with NA
cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
df[cols] = df[cols].replace(0, pd.NA)

print("\nMissing Values After Handling:")
print(df.isnull().sum())
# Histogram 1: Glucose
plt.figure()
plt.hist(df["Glucose"].dropna(), bins=20)
plt.title("Glucose Distribution")
plt.xlabel("Glucose")
plt.ylabel("Frequency")
plt.show()
# Histogram 2: BMI
plt.figure()
plt.hist(df["BMI"].dropna(), bins=20)
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()

# Boxplot 1: Age
plt.figure()
plt.boxplot(df["Age"].dropna())
plt.title("Age Distribution")
plt.ylabel("Age")
plt.show()
# Boxplot 2: Insulin
plt.figure()
plt.boxplot(df["Insulin"].dropna())
plt.title("Insulin Distribution")
plt.ylabel("Insulin Level")
plt.show()
# Correlation Analysis
df_clean = df.dropna()
correlation = df_clean.corr()
print("\nCorrelation Matrix:")
print(correlation)

print("\nCorrelation with Outcome:")
print(correlation["Outcome"].sort_values(ascending=False))


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("NAME:Piriyadharshini N\nROLL NO:24BAD086")
df = pd.read_csv("housing.csv")

print(df.head())
print(df.tail())
print(df.info())
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())

plt.scatter(df["area"], df["price"])
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()

corr = df.select_dtypes(include=["number"]).corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
#Trend interpretation
print(corr["price"].sort_values(ascending=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

print("NAME:Piriyadharshini N\nROLL NO:24BAD086")

df = pd.read_csv("marketing_campaign.csv", sep='\t')

print("\nFirst 5 rows")
print(df.head())

print("\nLast 5 rows")
print(df.tail())

print("\nData information")
print(df.info())

print("\nMissing values")
print(df.isnull().sum())

df['Age'] = 2024 - df['Year_Birth']
#barplot
print("\nAge Distribution")
age_counts = df["Age"].value_counts().sort_index()
plt.figure()
plt.bar(age_counts.index, age_counts.values)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
#boxplot
print("\nIncome Distribution")
df.boxplot(column='Income')
plt.title("Income Distribution")
plt.show()

print("\nSpending Pattern")
df[['MntWines', 'MntMeatProducts']].plot(kind='box')
plt.title("Customer Spending Pattern")
plt.show()
```
